# CGMD

* __NOTE__ `time-step = 15 fs` . See `cg_model/src/dynam_multithread.py`

* How to load CG files

```bash
pymol 1yrf_chain_a.pdb 1yrf.cor 1yrf.psf
```

* Load `dcd` trajectory file. In the Pymol console `load_traj output3.dcd`

* __NOTE__ `.dcd` files are trajectory files in binary format. This format is used by popular MD simulation packages such as `CHARMM`, `NAMD`, and `LAMMPS`

* Use `show sticks` to visualise a backbone linking all CG `CA` atoms. And to better visualize the changes in this backbone between the first and the last frame, do `hide spheres`
